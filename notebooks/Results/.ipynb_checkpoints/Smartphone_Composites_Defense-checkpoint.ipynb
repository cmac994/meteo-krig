```python
#Retrieve phase vector information for composite features from dataset
vec_bsteer_pos = np.radians(ds_comp_smart_pos['phase_arr'].values[:,4])
vec_bsteer_neg = np.radians(ds_comp_smart_neg['phase_arr'].values[:,4])
```

```python
#Composite phase for positive and negative feature composites 
top = 0; bot = 0
for i in range(0,len(vec_bsteer_pos)):
    top += np.sin(vec_bsteer_pos[i])
    bot += np.cos(vec_bsteer_pos[i])
    
vec_mean_pos = (np.pi/2)-np.arctan2(top,bot)
print(np.degrees(vec_mean_pos))

top = 0; bot = 0
for i in range(0,len(vec_bsteer_neg)):
    top += np.sin(vec_bsteer_neg[i])
    bot += np.cos(vec_bsteer_neg[i])
    
vec_mean_neg = (np.pi/2)-np.arctan2(top,bot)
print(np.degrees(vec_mean_neg))
```

```python
#Retrieve the mean spatial composite of feature surface variables (at T=0), i.e. center of time-series from previous figure 
ds_mean_pos = xr.open_dataset('../../data/Features/Composites/smart_full_meso_positive_spatialtemporal_cent_mean.nc')
ds_mean_neg = xr.open_dataset('../../data/Features/Composites/smart_full_meso_negative_spatialtemporal_cent_mean.nc')

#Import colortables
import cmasher as cmr
#Set minimum font size for plot
from metpy.plots import colortables
# READ AND PARSE HERE
ctable1 = 'NWSReflectivity'

#Assign colorbars 
import colorcet as cc
norm, cmapp_radar = colortables.get_with_steps(ctable1, 16, 16)
cmapp = cc.cm.rainbow_bgyrm_35_85_c71

#Initalize figure
fig = plt.figure(figsize=(20,25))
#Set default font size
fs=20
matplotlib.rcParams.update({'font.size': fs})

#Define spatial grid for composites (50 x 50 grid points => 250 x 250 km)
X,Y = np.meshgrid(np.arange(-25,25,1)*5,np.arange(-25,25,1)*5)
#Set grid units with metpy
from metpy.units import units
dx = (X[:,:-1]*0 + 5)*1000 * units.meter
dy = (Y[:-1,:]*0 + 5)*1000 * units.meter     

#Set perturbation range and remove index closest to zero
ns = np.arange(-1.2,1.2,0.2)
bidx = np.argwhere(abs(ns)<0.01).T[0]
ns = np.delete(ns,bidx)

#Retrieve vector components for feature normal wind (sample at reduced resolution for plotting)
nx = 5
Xskp = X[::nx,::nx]; Yskp = Y[::nx,::nx]
Upos = np.cos(vec_mean_pos)*ds_mean_pos['ustar_meso'].values[::nx,::nx]
Vpos = np.sin(vec_mean_pos)*ds_mean_pos['ustar_meso'].values[::nx,::nx]

Uneg = np.cos(vec_mean_neg)*ds_mean_neg['ustar_meso'].values[::nx,::nx]
Vneg = np.sin(vec_mean_neg)*ds_mean_neg['ustar_meso'].values[::nx,::nx]

#Compute the divergence/convergence of the perturbation feature normal wind 
from metpy import calc
div_pos = calc.divergence(np.cos(vec_mean_pos)*ds_mean_pos['ustar_meso'].values,np.sin(vec_mean_pos)*ds_mean_pos['ustar_meso'].values,dx, dy)
div_neg = calc.divergence(np.cos(vec_mean_neg)*ds_mean_neg['ustar_meso'].values,np.sin(vec_mean_neg)*ds_mean_neg['ustar_meso'].values,dx, dy)

#Define contour limits for divergence/convergence 
clev = np.arange(-2.5,2.75,0.25)
clev = clev[abs(clev)>1e-8]

#Multiply units for plot readability
div_pos = div_pos.magnitude*1e5
div_neg = div_neg.magnitude*1e5

#Mask out divergence/convergence where values are minimal
div_pos = np.ma.masked_where(abs(div_pos)<0.25,div_pos)
div_neg = np.ma.masked_where(abs(div_neg)<0.25,div_neg)

#Plot and contour perturbation pressure 
ax1 = plt.subplot(521)
plt.title("Pressure ($hPa$)",fontsize=fs+2)
plt.grid(True)
#Add row label
ax1.text(-0.25, 0.5, 'a)',rotation=0,
        verticalalignment='center', horizontalalignment='center',
        transform=ax1.transAxes,
        color='k', fontsize=fs+2)
im=plt.imshow(ds_mean_pos['p_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=cmr.prinsenvlag_r,vmin=-1.25,vmax=1.25)
CB=plt.contour(X,Y,ds_mean_pos['p_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.ylabel('(Y) N-S (km)')
plt.colorbar(im)
#Add perturbation feature normal wind vectors
q1 = ax1.quiver(Xskp,Yskp,Upos,Vpos,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax1.contour(X,Y,div_pos,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax1.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
ax2 = plt.subplot(522)
plt.title("Pressure ($hPa$)",fontsize=fs+2)
im=plt.imshow(ds_mean_neg['p_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=cmr.prinsenvlag_r,vmin=-1.25,vmax=1.25)
CB=plt.contour(X,Y,ds_mean_neg['p_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
#Add perturbation feature normal wind vectors
q2 = ax2.quiver(Xskp,Yskp,Uneg,Vneg,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax2.contour(X,Y,div_neg,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax2.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.grid(True)
plt.colorbar(im)

#Plot and contour perturbation feature normal wind
ax3 = plt.subplot(523)
plt.title("Feature Normal Wind ($m/s$)",fontsize=fs+2)
plt.grid(True)
plt.ylabel('(Y) N-S (km)')
#Add row label
ax3.text(-0.25, 0.5, 'b)',rotation=0,
        verticalalignment='center', horizontalalignment='center',
        transform=ax3.transAxes,
        color='k', fontsize=fs+2)
im=plt.imshow(ds_mean_pos['ustar_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.PRGn,vmin=-1,vmax=1)
CB=plt.contour(X,Y,ds_mean_pos['ustar_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.colorbar(im)
#Add perturbation feature normal wind vectors
q3 = ax3.quiver(Xskp,Yskp,Upos,Vpos,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax3.contour(X,Y,div_pos,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax3.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
ax4 = plt.subplot(524)
plt.title("Feature Normal Wind ($m/s$)",fontsize=fs+2)
im=plt.imshow(ds_mean_neg['ustar_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.PRGn,vmin=-1,vmax=1)
CB=plt.contour(X,Y,ds_mean_neg['ustar_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.grid(True)
#Add perturbation feature normal wind vectors
q4 = ax4.quiver(Xskp,Yskp,Uneg,Vneg,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax4.contour(X,Y,div_neg,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax4.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.colorbar(im)

#Plot and contour perturbation temperature
ax5 = plt.subplot(525)
plt.title("Temperature ($\degree C$)",fontsize=fs+2)
plt.grid(True)
#Add row label
ax5.text(-0.25, 0.5, 'c)',rotation=0,
        verticalalignment='center', horizontalalignment='center',
        transform=ax5.transAxes,
        color='k', fontsize=fs+2)
im=plt.imshow(ds_mean_pos['t_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.RdBu_r,vmin=-0.5,vmax=0.5)
CB=plt.contour(X,Y,ds_mean_pos['t_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.ylabel('(Y) N-S (km)')
plt.colorbar(im)
#Add perturbation feature normal wind vectors
q5 = ax5.quiver(Xskp,Yskp,Upos,Vpos,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax5.contour(X,Y,div_pos,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax5.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
ax6 = plt.subplot(526)
plt.title("Temperature ($\degree C$)",fontsize=fs+2)
im=plt.imshow(ds_mean_neg['t_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.RdBu_r,vmin=-0.5,vmax=0.5)
CB=plt.contour(X,Y,ds_mean_neg['t_meso'].values,levels=ns,colors='k')
#Add perturbation feature normal wind vectors
q6 = ax6.quiver(Xskp,Yskp,Uneg,Vneg,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax6.contour(X,Y,div_neg,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax6.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.grid(True)
plt.colorbar(im)

#Define seperate contour levels for dew point due to reduced pertubation magnitude
ns = [-0.25,-0.2,-0.15,-0.1,-0.05,0.05,0.1,0.15,0.2,0.25]
#Plot and contour perturbation dew point temperature
ax7 = plt.subplot(527)
plt.title("Dew Point ($\degree C$)",fontsize=fs+2)
plt.grid(True)
#Add row label
ax7.text(-0.25, 0.5, 'd)',rotation=0,
        verticalalignment='center', horizontalalignment='center',
        transform=ax7.transAxes,
        color='k', fontsize=fs+2)
im=plt.imshow(ds_mean_pos['d_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.BrBG,vmin=-0.25,vmax=0.25)
CB=plt.contour(X,Y,ds_mean_pos['d_meso'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.colorbar(im)
plt.ylabel('(Y) N-S (km)')
#Add perturbation feature normal wind vectors
q7 = ax7.quiver(Xskp,Yskp,Upos,Vpos,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax7.contour(X,Y,div_pos,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax7.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
ax8 = plt.subplot(528)
plt.title("Dew Point ($\degree C$)",fontsize=fs+2)
im=plt.imshow(ds_mean_neg['d_meso'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=plt.cm.BrBG,vmin=-0.25,vmax=0.25)
CB=plt.contour(X,Y,ds_mean_neg['d_meso'].values,levels=ns,colors='k')
#Add perturbation feature normal wind vectors
q8 = ax8.quiver(Xskp,Yskp,Uneg,Vneg,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax8.contour(X,Y,div_neg,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax8.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.grid(True)
plt.colorbar(im)

ns = [10,12,14,16,18,20,22,24]

#Plot and contour composite reflectivity
ax9 = plt.subplot(529)
plt.title("Composite Reflectivity ($dBZ$)",fontsize=fs+2)
plt.grid(True)
im=plt.imshow(ds_mean_pos['c_refl'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=cmapp_radar,vmin=0,vmax=35)
plt.colorbar(im)
#Add row label
ax9.text(-0.25, 0.5, 'e)',rotation=0,
        verticalalignment='center', horizontalalignment='center',
        transform=ax9.transAxes,
        color='k', fontsize=fs+2)
#CB=plt.contour(X,Y,ds_mean_pos['c_refl'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
plt.xlabel('(X) E-W (km)')
plt.ylabel('(Y) N-S (km)')
#Add perturbation feature normal wind vectors
q9 = ax9.quiver(Xskp,Yskp,Upos,Vpos,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax9.contour(X,Y,div_pos,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax9.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.ylabel('(Y) N-S (km)')
ax10 = plt.subplot(5,2,10)
plt.title("Composite Reflectivity ($dBZ$)",fontsize=fs+2)
im=plt.imshow(ds_mean_neg['c_refl'].values,aspect='auto',origin='lower',interpolation='nearest',extent=[-125,125,-125,125],cmap=cmapp_radar,vmin=0,vmax=35)
plt.colorbar(im)
plt.xlabel('(X) E-W (km)')
#CB=plt.contour(X,Y,ds_mean_neg['c_refl'].values,levels=ns,colors='k')
plt.clabel(CB,fmt='%1.2f',inline=1,fontsize=14)
#Add perturbation feature normal wind vectors
q10 = ax10.quiver(Xskp,Yskp,Uneg,Vneg,zorder=5,scale_units='xy',scale=0.04,width=0.0035)
#Contour and label divergence/convergence
CS = ax10.contour(X,Y,div_neg,levels=clev,colors='saddlebrown',alpha=1,linewidths=2,zorder=10)
ax10.clabel(CS, CS.levels, inline=True, fmt="%1.1f", fontsize=14, colors='saddlebrown') #Label contours
plt.grid(True)

plt.tight_layout()
plt.subplots_adjust(top=0.94)
plt.suptitle('Smartphone Pressure Feature Mean Composite Analysis (T=0): Positive (left) and Negative (right)',fontsize=fs+4)
```

```python
#Loop the first three columns of the beamsteering data array (containing composite phase vectors for different variables: 0=feature normal wind, 1=temperature, 2=dew point)
for i in range(0,3):
    #Compute the direction of phase propogation for positive composite features
    dd_pos = 90-(np.degrees(ds_comp_smart_pos['phi'].values[:,i]*-1))
    dd_pos[dd_pos<0] = dd_pos[dd_pos<0]+360
    dd_pos[dd_pos>360] = dd_pos[dd_pos>360]-360

    ds_comp_smart_pos['phi'].values[:,i] = dd_pos
    
    #Compute the direction of phase propogation for positive composite features
    dd_neg = 90-(np.degrees(ds_comp_smart_neg['phi'].values[:,i]*-1))
    dd_neg[dd_neg<0] = dd_neg[dd_neg<0]+360
    dd_neg[dd_neg>360] = dd_neg[dd_neg>360]-360
    
    ds_comp_smart_neg['phi'].values[:,i] = dd_neg
```

```python
#Import colormap and wind rose repositories
import cmasher as cmr
from windrose import WindroseAxes,WindAxes,WindAxesFactory

#Set default fontize for plot
fs=20
matplotlib.rcParams.update({'font.size': fs})

#Initialize plot
fig = plt.figure(figsize=(20,14))
#Retrieve magnitude squared coherence values (between given surface variable and perturbation pressure)
ds_comp_smart_pos['ccoh'].values
#Define colormaps for wind roses
import colorcet as cc
cmapp2 = cc.cm.rainbow_bgyrm_35_85_c71

binns = np.arange(0,1.1,0.1) #set bins for wind roses (0-1 for coherence)

#Plot phase propogation and magnitude squared coherence of positive perturbation pressure and perturbation feature normal wind
ax1 = fig.add_subplot(231,projection="windrose")
ax1.bar(ds_comp_smart_pos['phi'].values[:,0], ds_comp_smart_pos['ccoh'].values[:,0], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax1.set_title('a) Pressure & Feature Normal Wind',fontsize=fs+2)
label_position=ax1.get_rlabel_position()
ax1.text(np.radians(label_position+10),ax1.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax1.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])
#Add row title
ax1.text(-0.15, 0.5, 'Positive',rotation=90,
        verticalalignment='center', horizontalalignment='center',
        transform=ax1.transAxes,
        color='k', fontsize=fs+4)

#Plot phase propogation and magnitude squared coherence of positive perturbation pressure and perturbation temperature
ax2 = fig.add_subplot(232,projection="windrose")
ax2.bar(ds_comp_smart_pos['phi'].values[:,1], ds_comp_smart_pos['ccoh'].values[:,1], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax2.set_title('b) Pressure & Temperature',fontsize=fs+2)
ax2.text(np.radians(label_position+10),ax2.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax2.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])

#Plot phase propogation and magnitude squared coherence of positive perturbation pressure and perturbation dew point temperature
ax3 = fig.add_subplot(233,projection="windrose")
ax3.bar(ds_comp_smart_pos['phi'].values[:,2], ds_comp_smart_pos['ccoh'].values[:,2], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax3.set_title('c) Pressure & Dew Point',fontsize=fs+2)
ax3.text(np.radians(label_position+10),ax3.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax3.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])

#Plot phase propogation and magnitude squared coherence of perturbation pressure and perturbation feature normal wind
ax4 = fig.add_subplot(234,projection="windrose")
ax4.bar(ds_comp_smart_neg['phi'].values[:,0], ds_comp_smart_neg['ccoh'].values[:,0], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax4.set_title('d) Pressure & Feature Normal Wind',fontsize=fs+2)
label_position=ax1.get_rlabel_position()
ax4.text(np.radians(label_position+10),ax4.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax4.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])
#Add row title
ax4.text(-0.15, 0.5, 'Negative',rotation=90,
        verticalalignment='center', horizontalalignment='center',
        transform=ax4.transAxes,
        color='k', fontsize=fs+4)

#Plot phase propogation and magnitude squared coherence of perturbation pressure and perturbation feature normal wind
ax5 = fig.add_subplot(235,projection="windrose")
ax5.bar(ds_comp_smart_neg['phi'].values[:,1], ds_comp_smart_neg['ccoh'].values[:,1], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax5.set_title('e) Pressure & Temperature',fontsize=fs+2)
ax5.text(np.radians(label_position+10),ax5.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax5.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])

#Plot phase propogation and magnitude squared coherence of perturbation pressure and perturbation feature normal wind
ax6 = fig.add_subplot(236,projection="windrose")
l6 = ax6.bar(ds_comp_smart_neg['phi'].values[:,2], ds_comp_smart_neg['ccoh'].values[:,2], bins=binns, normed=True, opening=0.8, edgecolor='white',cmap=cmapp2)
#Add title and plot labels
ax6.set_title('f) Pressure & Dew Point',fontsize=fs+2)
ax6.text(np.radians(label_position+10),ax6.get_rmax()*0.75,'Frequency',
    rotation=label_position,ha='center',va='center')
ax6.set_xticklabels(['0', '$\pi$/4', '$\pi$/2', '3$\pi$/2','-$\pi$','-3$\pi$/2','-$\pi$/2','-$\pi$/4'])

#Plot legend and adjust subplot to add subplot title
plt.tight_layout(h_pad=2.5)
lg1 = plt.legend( handles=[l6], labels=['10:20','20:30','30:40','40:50'], loc='lower center', bbox_to_anchor = (0,-0.07,1,1), ncol=6, fontsize=18,fancybox=True,
        bbox_transform = plt.gcf().transFigure )

plt.subplots_adjust(top=0.87)
plt.suptitle('Smartphone Pressure Feature Phase and Coherence Distribution: Positive (top) and Negative (bottom)',fontsize=fs+4)
```
